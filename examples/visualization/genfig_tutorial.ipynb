```python
insertion = genfig.Graph(db='.gen/default.db', block_group_name='insertion' )
deletion = genfig.Graph(db='.gen/default.db', block_group_name='deletion' )

# The Graph object itself can already be displayed with a default block layout
display(insertion)
display(deletion)
```

```python
# Renders can be customized using the render_graph and render_block_graph methods.

# They return a png bytestring that can be displayed using IPython.display.Image

from IPython.display import Image

display(Image(deletion.render_graph(minimize=True, format='png')))
display(Image(insertion.render_graph(minimize=True, format='png')))

```

```python
# Repeat the above plots but save as DOT instead

insertion.render_graph(minimize=True, format='dot', filename='insertion.dot')
deletion.render_graph(minimize=True, format='dot', filename='deletion.dot')

insertion.render_block_graph(minimize=True, format='dot', filename='insertion_blocks.dot')
deletion.render_block_graph(minimize=True, format='dot', filename='deletion_blocks.dot')
```
